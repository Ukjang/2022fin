```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
game = True
size_tf = True
player = 0
p1_all_list = []
p1_x_list = []
p1_y_list = []
p2_all_list = []
p2_x_list = []
p2_y_list = []
p_all = []
idx_list = [-4,-3,-2,-1, 0, 1, 2, 3, 4]
p1_score_list = []
p2_score_list = []
cross_list = [(1,1), (1, -1), (-1, 1), (-1, -1)]

while size_tf :
  try :
    om_size = int(input('판크기를 입력해주세요 :'))
    size_tf = False
    break
  except:
    print('자연수를 입력해주세요\n')
    continue

while game :
  if player % 2 == 0 :
    while True:
      try:
        p1 = input('첫번째 플레이어의 위치를 입력해주세요: ex: 5 5 :')
        p1_x = int(str(p1).strip().split(' ')[0])
        p1_y = int(str(p1).strip().split(' ')[-1])
        if p1_x >= om_size or p1_y >= om_size:
          print('바둑판을 벗어났습니다.\n')
          continue
        if (p1_x, p1_y) in p_all:
          print('이미 바둑돌이 놓여 있습니다\n')
          continue


        break
      except :
        print('ex: 5 5 형식으로 입력해주세요\n')
      
    
    p1_x_list.append(p1_x)
    p1_y_list.append(p1_y)
    p1_all_list.append((p1_x, p1_y))
    p_all.append((p1_x, p1_y))
    print(f'첫번째 플레이어 바둑돌 위치: {p1_x}, {p1_y}')
  elif player % 2 == 1 :
    while True:
      try:
        p2 = input('두번째 플레이어의 위치를 입력해주세요: ex: 5 5 :')
        p2_x = int(str(p2).strip()[0])
        p2_y = int(str(p2).strip()[-1])
        if p2_x >= om_size or p2_y >= om_size:
          print('바둑판을 벗어났습니다.\n')
          continue
        if (p2_x, p2_y) in p_all:
          print('이미 바둑돌이 놓여 있습니다\n')
          continue
        break
      except :
        print('ex: 5 5 형식으로 입력해주세요\n')
      
    
    p2_x_list.append(p2_x)
    p2_y_list.append(p2_y)
    p2_all_list.append((p2_x, p2_y))
    p_all.append((p2_x, p2_y))
    print(f'두번째 플레이어 바둑돌 위치: {p2_x}, {p2_y}')
  
  
  plt.figure(figsize=(3, 3))
  plt.scatter(p1_x_list, p1_y_list, label='player1')
  plt.scatter(p2_x_list, p2_y_list, label='player2')
  plt.xticks(range(om_size))
  plt.yticks(range(om_size))
  plt.legend()
  plt.grid()
  plt.show()
  

  # 승리조건 
  if player % 2 == 0 :
    # 가로 5개
    for i in [1, -1]:
        if (p1_x + i, p1_y) in p1_all_list:
          for idx in range(5):
            w_temp = idx_list[idx: idx + 5]
            w_p1_score = 0
            for n in w_temp:
              if (p1_x + (i * n), p1_y) in p1_all_list:
                w_p1_score += 1
            p1_score_list.append(w_p1_score)
        
    # 세로 5개
    for i in [1, -1]:
        if (p1_x, p1_y + i) in p1_all_list:
          for idx in range(5):
            w_temp = idx_list[idx: idx + 5]
            w_p1_score = 0
            for n in w_temp:
              if (p1_x , p1_y + (i * n)) in p1_all_list:
                w_p1_score += 1
            p1_score_list.append(w_p1_score)
        
    # 대각선
    for (i, j) in cross_list:
        if (p1_x + i, p1_y + j) in p1_all_list:
          for idx in range(5):
            w_temp = idx_list[idx: idx + 5]
            w_p1_score = 0
            for n in w_temp:
              if (p1_x + (i * n), p1_y + (j * n)) in p1_all_list:
                w_p1_score += 1
            p1_score_list.append(w_p1_score)
    try:
        if max(p1_score_list) == 5:
            print('첫번째 플레이어가 승리하였습니다')  
            print('게임이종료되었습니다.')
            game = False
    except:
        pass    
  elif player % 2 == 1 :
  # 가로 5개
    for i in [1, -1]:
        if (p2_x + i, p2_y) in p2_all_list:
          for idx in range(5):
            w_temp = idx_list[idx: idx + 5]
            w_p2_score = 0
            for n in w_temp:
              if (p2_x + (i * n), p2_y) in p2_all_list:
                w_p2_score += 1
            p2_score_list.append(w_p2_score)
        
    # 세로 5개
    for i in [1, -1]:
        if (p2_x, p2_y + i) in p2_all_list:
          for idx in range(5):
            w_temp = idx_list[idx: idx + 5]
            w_p2_score = 0
            for n in w_temp:
              if (p2_x , p2_y + (i * n)) in p2_all_list:
                w_p2_score += 1
            p2_score_list.append(w_p2_score)
        
    # 대각선
    for (i, j) in cross_list:
        if (p2_x + i, p2_y + j) in p2_all_list:          
          for idx in range(5):
            w_temp = idx_list[idx: idx + 5]
            w_p2_score = 0
            for n in w_temp:
              if (p2_x + (i * n), p2_y + (j * n)) in p2_all_list:
                w_p2_score += 1
            p2_score_list.append(w_p2_score)
    try :
        if max(p2_score_list) == 5:
            print('두번째 플레이어가 승리하였습니다')  
            print('게임이종료되었습니다.')
            game = False
    except:
        pass

  if (len(p1_all_list) + len(p2_all_list)) == om_size * om_size :
    print('바둑판이 다 찼습니다. 무승부')
    print('게임이종료되었습니다.')
    game = False
    
  player += 1
```
