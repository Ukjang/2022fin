```python
PDF_DIR = './실습1/pdf_files/'
from tqdm.notebook import tqdm
import shutil
file_list = os.listdir(PDF_DIR )
file_list
for file in tqdm(file_list):
    if file[-3:] == 'pdf':
        date = file.split('_')[1]
        if os.path.exists(PDF_DIR + date) != True :
            os.makedirs(PDF_DIR + date)
        else :
            pass
        shutil.copyfile(PDF_DIR + file, PDF_DIR + date +'/' + file)
```
