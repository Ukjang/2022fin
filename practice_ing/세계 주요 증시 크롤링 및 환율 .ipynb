# 세계 주요 증시 크롤링

```python
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import time
from tqdm.notebook import tqdm
```

```python


url = 'https://finance.naver.com/world/'
html = requests.get(url)
soup = bs(html.text, 'html.parser')
soup
```

```python


driver = webdriver.Chrome('/pydata/Chromedriver.exe')
url = 'https://finance.naver.com/world/'
driver.get(url)
driver.implicitly_wait(10)
soup = bs(driver.page_source, 'html.parser')
soup
```

```python
soup.select_one('.section_quot')
```

```python
soup.select_one('.section_quot').select('.point_up')
```

```python
soup.select_one('.section_quot').select('.point_up')[0].select('td')
```

```python
[td.text.strip() for td in soup.select_one('.section_quot').select('.point_up')[0].select('td')]
```

```python
for region in soup.select_one('.section_quot').select('.point_up'):
    print([td.text.strip() for td in region.select('td')])
```
