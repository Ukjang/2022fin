```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.jungle.co.kr/'
html = requests.get(url)
soup = bs(html.text, 'html.parser')
soup
```

```python
result_list = []
for thumb in soup.select('.thumb_list'):
    for li in thumb.select('li'):
        try :
            title = li.select_one('.title').text
            cate = li.select('a')[-1].text.strip()
            temp = {'제목' : title,
                   '분류' : cate}
            print(temp)
            result_list.append(temp)
        except :
            pass
```
