```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
from tqdm.notebook import tqdm
import time
chrome_path = r'C:\pydata\chromedriver.exe'
driver = webdriver.Chrome(chrome_path)
url = 'https://www.jungle.co.kr/'
driver.get(url)
for _ in tqdm(range(1, 150)):
    driver.find_element_by_css_selector('#more').click()
    driver.implicitly_wait(1)
    time.sleep(0.5)
html = driver.page_source
soup = bs(html, 'html.parser')
result_list = []
for thumb in soup.select('.thumb_list'):
    for li in thumb.select('li'):
        title = li.select_one('.title').text
        cate = li.select_one('.thumb_cate').text
        temp = {'제목': title,
               '분류' : cate}
        if temp not in result_list :
            print(temp)
            result_list.append(temp)
```

```python
print('총개수 : ',len(result_list))
```

```python
el_list = driver.find_elements(By.CLASS_NAME, 'thumb_list')
for el in el_list :
    print(el.text)
```

```python
elements =driver.find_elements(By.TAG_NAME , 'li')
for el in elements:
    print(el.text, '\n')
```

## 문제 2

```python
from selenium import webdriver
import time
driver = webdriver.Chrome('/pydata/chromedriver.exe')
result_list = []
for pg in range(1, 11) :
    url = f'https://www.rocketpunch.com/jobs?page={pg}'
    driver.get(url)
    driver.implicitly_wait(3)
    time.sleep(3)
    html = driver.page_source
    soup = bs(html, 'html.parser')
    for lst in soup.select('.company.item') :
        c_name = lst.select_one('.header.name').text.strip().replace('\n', '')
        c_int = lst.select_one('.description').text.strip()
        for noti in lst.select('a[class *= job-title]'):
            c_noti = noti.text
            print('회사명: ', c_name, '회사소개:', c_int, '공고: ', c_noti, '\n')
            result_list.append({'회사명': c_name, 
                                '회사소개': c_int, 
                                '공고': c_noti})
    time.sleep(0.5)
len(result_list)
```
